```python
import tensorflow as tf

x = tf.constant(3)
y = tf.constant(5)
final_node = tf.add(x, y)

result = final_node.numpy()
print(result, type(result))
```

```python
import tensorflow as tf

# Computational Graph:
c1 = tf.constant(0.034, dtype=tf.float64)
c2 = tf.constant(1000.0, dtype=tf.float64)
x = tf.multiply(c1, c1)
y = tf.multiply(c1, c2)
final_node = tf.add(x, y)

# No need for explicit session management, just run the final node
result = final_node.numpy()
print(result, type(result))
```
