```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for learning
np.random.seed(0)  # for reproducibility
num_samples = 100
learn_data = np.random.randn(num_samples, 2)  # Generate random data points with 2 features
learn_labels = np.random.randint(0, 2, num_samples)  # Generate random labels (0 or 1)

# Generate random data for testing
test_data = np.random.randn(num_samples, 2)  # Generate random test data points with 2 features
test_labels = np.random.randint(0, 2, num_samples)  # Generate random test labels (0 or 1)

# Print the shapes to verify
print("learn_data shape:", learn_data.shape)
print("learn_labels shape:", learn_labels.shape)
print("test_data shape:", test_data.shape)
print("test_labels shape:", test_labels.shape)

fig, ax = plt.subplots()

# Plotting learn data
colours = ('green', 'blue')
if learn_data is not None and learn_labels is not None:  # Check if learn_data and learn_labels are not None
    for n_class in range(2):
        ax.scatter(learn_data[learn_labels==n_class][:, 0],
                   learn_data[learn_labels==n_class][:, 1],
                   c=colours[n_class], s=40, label=str(n_class))

# Plotting test data
colours = ('lightgreen', 'lightblue')
if test_data is not None and test_labels is not None:  # Check if test_data and test_labels are not None
    for n_class in range(2):
        ax.scatter(test_data[test_labels==n_class][:, 0],
                   test_data[test_labels==n_class][:, 1],
                   c=colours[n_class], s=40, label=str(n_class))

# Generating a linear model (example)
class LinearModel:
    def __init__(self, num_features):
        self.weights = np.random.randn(num_features + 1)  # Additional weight for bias term

# Creating a linear model with 2 features (plus one bias term)
p = LinearModel(num_features=2)

# Plotting the line
samples = learn_data
if samples is not None and p is not None:  # Check if samples and p are not None
    X = np.arange(np.max(samples[:,0]))
    m = -p.weights[0] / p.weights[1]
    c = -p.weights[2] / p.weights[1]
    ax.plot(X, m * X + c)

plt.show()


```
